### Analyzing and Observing the sales, discount and profit in different region

```python
region_profit = myData.groupby('Region')['Profit'].sum().sort_values(ascending = True)
region_sales = myData.groupby('Region')['Sales'].sum().sort_values(ascending = True)
region_discount = myData.groupby('Region')['Discount'].sum().sort_values(ascending = True)
region_sales.plot(kind='bar',color='#b30059',title='Observing the sales in different region')
plt.show()
region_discount.plot(kind='bar',color='#5c0099',title='Observing the discount in different region')
plt.show()
region_profit.plot(kind='bar',color='#004d4d',title='Observing the profit in different region')
plt.show()
```

### Analyzing and Observing the sales, discount and profit in different ship-mode

```python
shipMode_profit = myData.groupby('Ship Mode')['Profit'].sum().sort_values(ascending = True)
shipMode_sales = myData.groupby('Ship Mode')['Sales'].sum().sort_values(ascending = True)
shipMode_discount = myData.groupby('Ship Mode')['Discount'].sum().sort_values(ascending = True)
shipMode_sales.plot(kind='bar',color='#e60073',title='Observing the sales in different ship mode')
plt.show()
shipMode_discount.plot(kind='bar',color='#7a00cc',title='Observing the discount in different ship mode')
plt.show()
shipMode_profit.plot(kind='bar',color='#006666',title='Observing the profit in different ship mode')
plt.show()
```
